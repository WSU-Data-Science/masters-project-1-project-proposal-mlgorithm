```python
import numpy as np
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
img=plt.imread('chess.jpg')
plt.imshow(img)
print(img.shape)
```

```python
input_arr=np.reshape(img, (1,img.shape[0]*img.shape[1]*img.shape[2]))
print(input_arr.shape)
print(225*225*3)
X=np.zeros((input_arr.shape[1],3))
y=np.zeros((input_arr.shape[1]))
X.shape
y.shape
index=0
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        for k in range(img.shape[2]):
            X[index,0]=i
            X[index,1]=j
            X[index,2]=i
            y[index]=img[i,j,k]
            index+=1
```

```python
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.3)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
model=MLPClassifier(hidden_layer_sizes=(50, 30))
```

```python
model.fit(X_train, y_train)
```

```python
y_pred =model.predict(X)
```

```python
def eval(y_test, y_pred): 
    TP=0
    TN=0
    FP=0
    FN=0
    for i in range(y_test.shape[0]): 
        if(y_test[i]==y_pred[i]): 
            if(y_pred[i]==1):
                TP+=1
            else:
                TN+=1
        else:
            if(y_pred[i]==1):
                FP+=1
            else:
                FN+=1
    return(TP, TN, FP, FN)
```

```python
TP,TN, FP, FN = eval(y,y_pred)
```

```python
acc = (TP + TN)/(TP+TN + FP+ FN)
acc
```

```python
img_out =np.reshape(y_pred,(img.shape[0],img.shape[1],img.shape[2]))

plt.imshow(img_out)
```
